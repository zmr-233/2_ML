### |DQN-基于价值 & 策略梯度-基于策略
Q-learning、DQN 及 DQN 改进算法都是**基于价值（value-based）**的方法，而 **基于策略（policy-based）** 的方法：
策略梯度是基于策略的方法的基础，本章从策略梯度算法说起。

## 策略梯度算法
目标策略$\pi_\theta$是一个随机性策略，可以用一个线性模型/神经网络来为策略函数建模。目标是要寻找一个最优策略并最大化这个策略的期望回报，目标函数为：
$$
J(\theta)=\mathbb{E}_{s_{0}}\left[V^{\pi_{\theta}}\left(s_{0}\right)\right]
$$
其中$s_0$为初始状态，将目标函数J对$\theta$求导，就可以用梯度上升来最大化目标函数

对目标函数求梯度，可以得到如下式子:
$$
\begin{aligned} \nabla_{\theta} J(\theta) & \propto \sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \pi_{\theta}(a \mid s) \\ & =\sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} \pi_{\theta}(a \mid s) Q^{\pi_{\theta}}(s, a) \frac{\nabla_{\theta} \pi_{\theta}(a \mid s)}{\pi_{\theta}(a \mid s)} \\ & =\mathbb{E}_{\pi_{\theta}}\left[Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \log \pi_{\theta}(a \mid s)\right]\end{aligned}
$$
1. **目标函数**:$J(\theta)$是策略$\pi_{\theta}$的性能度量,表示该策略的期望回报
2. **策略梯度**:$\nabla_{\theta} J(\theta)$
3. **状态访问频率**:$\nu^{\pi_{\theta}}(s)$表示该策略下访问状态s的频率
4. **动作价值函数**:$Q^{\pi_{\theta}}(s, a)$表示s,a获得的期望回报
5. **策略梯度的推导**:
   - 首先，梯度是通过所有状态的加权和来计算的，权重是每个状态的访问频率
   - 对于每个状态，我们考虑所有可能的动作价值，并乘以算子
   - 最后，我们将这个求和转化为一个期望值
   
> 注意：策略梯度为在线策略算法!因为上式中期望$\mathbb{E}_{\pi_{\theta}}$的下标是$\pi_\theta$，即必须使用当前策略$\pi_\theta$采样的数据来计算梯度

---

### REINFORCE 算法
可以用多种方式估计$Q^{\pi_{\theta}}(s, a)$，接下来要介绍的 REINFORCE 算法便是采用了蒙特卡洛方法来估计。于一个有限步数的环境来说，REINFORCE 算法中的策略梯度为：
$$
\nabla_{\theta} J(\theta)=\mathbb{E}_{\pi_{\theta}}\left[\sum_{t=0}^{T}\left(\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}\right) \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)\right]
$$
其中，T是和环境交互的最大步数。例如，在车杆环境中，T=200

#### |REINFORCE具体实现流程：
- 初始化策略参数$\theta$
- **for** 序列$e=1 \rightarrow E$ **do** :
- ----用当前策略$\pi_{\theta}$采样轨迹$\left\{s_{1}, a_{1}, r_{1}, s_{2}, a_{2}, r_{2}, \ldots s_{T}, a_{T}, r_{T}\right\}$
- ----计算当前轨迹每个时刻t往后的回报$\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}$ ，记为 $\psi_{t}$
- ----对$\theta$进行更新,$\theta=\theta+\alpha \sum_{t}^{T} \psi_{t} \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$
- **end for**

### 策略网络
这里采用在离散动作空间上的softmax()函数来实现一个可学习的**多项分布（multinomial distribution）**

#### softmax()激活函数
将任意实数向量转换为概率分布。给定一个实数向量，Softmax将为每个元素分配一个介于0和1之间的概率值，且所有元素的概率值之和为1。
$$
\text{Softmax}(z)_i = \frac{e^{z_i}}{\sum_j e^{z_j}}
$$
在策略网络中，使用Softmax是很有意义的，因为我们想要输出的是一个概率分布（即每个动作的概率）。Softmax确保了输出是一个有效的概率分布。

但是，ReLU不适合用作策略网络的输出层的激活函数，因为它不会输出一个概率分布。ReLU的输出是非负的，但它们的和不一定是1，而且它们也不一定都在0和1之间

````python
class PolicyNet(torch.nn.Module):
    def __init__(self, state_dim, hidden_dim, action_dim):
        super(PolicyNet, self).__init__()
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, action_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        #x = F.relu(self.fc2(x))
        x = self.fc2(x)
        return F.softmax(x, dim=1)  # 输出各动作的概率
````

### REINFORCE算法
按照算法将损失函数写为策略回报的负数，$-\sum_{t} \psi_{t} \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$，对$\theta$求导就可以通过梯度下降来更新策略


````python
class REINFORCE:
    def __init__(self, state_dim, hidden_dim, action_dim, learning_rate, gamma,
                 device):
        self.policy_net = PolicyNet(state_dim, hidden_dim,
                                    action_dim).to(device)
        self.optimizer = torch.optim.Adam(self.policy_net.parameters(), lr=learning_rate)  # 使用Adam优化器
        self.gamma = gamma  # 折扣因子
        self.device = device
    def take_action(self, state):  # 根据动作概率分布随机采样
        state = torch.tensor(state, dtype=torch.float).to(self.device)
        probs = self.policy_net(state.unsqueeze(0))  
        # ---------------给state加一个维度👆
            #网络通常期望的输入是一个批次的数据，即为[batch_size, state_dim]
            #即使只有一个状态，也需要视作批次大小为1的批次
            #[0.5, 1.2, -0.3]->[[0.5, 1.2, -0.3],] batch_size = 1
            #为了高效地进行计算，通常一次处理多个数据样本
        action_dist = torch.distributions.Categorical(probs)  # 根据probs给出的概率创建一个类别分布
        #---------------------使用Categorical分布👆
            #是离散概率分布的一种，描述具有有限数量可能结果的随机实验的概率分布
        action = action_dist.sample()  # 从创建的类别分布中抽样, 得到某个类别的序号
        return action.item()

    def update(self, transition_dict):
        return_list = transition_dict['rewards']
        state_list = transition_dict['states']
        action_list = transition_dict['actions']

        G = 0
        self.optimizer.zero_grad() # 设定0梯度
        for i in reversed(range(len(return_list))):  # 从最后一步算起, 对每一个状态动作计算损失
            reward = return_list[i]
            state = torch.tensor(state_list[i], dtype=torch.float).to(self.device)
            action = torch.tensor(action_list[i]).view(-1, 1).to(self.device)  # 选择当前步骤的动作
            
            log_prob = torch.log(
                self.policy_net(state.unsqueeze(0)).gather(1, action))  # 依据动作选概率, 对数化
            
            G = self.gamma * G + reward
            loss = -log_prob * G  # 每一步的损失函数
            loss.backward()  # 反向传播计算梯度, 在循环中一直积累梯度
        self.optimizer.step()  # 梯度下降, 每轮游戏结束后计算的总损失梯度全部更新一次
````

#### update与公式的关系：
如上update与公式$\nabla_{\theta} J(\theta)=\mathbb{E}_{\pi_{\theta}}\left[\sum_{t=0}^{T}\left(\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}\right) \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)\right]$的对应关系：

> 注意：梯度是靠loss.backward()自动计算的，这是通过链式法则完成的！！！

1. **期望计算**:
   - 公式中的期望$\mathbb{E}_{\pi_{\theta}}[\cdot]$表示按策略$\pi_{\theta}$ 采样的轨迹的期望。在`update`函数中，没有显式地计算期望，而是通过实际的轨迹（即从实际的环境交互中采样的状态、动作和奖励序列）来进行更新
2. **奖励的累积**:
   - 公式中的$\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}$ 表示从时间步$t$到结束的累积折扣奖励。在`update`函数中，这通过变量 `G` 来计算，它在每次循环迭代中都会累积（折扣的）奖励。
3. **策略的对数梯度**:
   - 公式中的$\nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$是在时间步$t$，给定状态$s_t$和采取的动作$a_t$下策略的对数梯度。在`update`函数中，这通过以下代码来计算：
     ```python
     log_prob = torch.log(self.policy_net(state.unsqueeze(0)).gather(1, action))
     ```
4. **损失函数**:
   - 在策略梯度方法中，我们的目标是最大化期望的累积奖励。但是，由于优化器通常最小化损失，所以我们使用负的策略梯度作为损失：
     ```python
     loss = -log_prob * G
     ```
5. **梯度的累积**:
   - 对于每一个时间步，都计算损失并调用`loss.backward()`，这会累积梯度，直到整个轨迹都处理完。
6. **参数更新**:
   - 使用`self.optimizer.step()`，我们使用累积的梯度来更新策略参数，实现了策略梯度的更新。

### ✅注意点
策略梯度的损失，不是与正确的动作来对比得到交叉熵，因为我们事先不知道正确的动作，所以没有参考，因此就只能人为确定一种损失，例如用网络给出的各个动作的概率，乘以当前得到的奖励，如果奖励比较低，乘起来也比较小，我们希望提升奖励，所以加个负号再执行梯度下降，等于是做了梯度上升，也就希望提升奖励，因为概率肯定是正的，所以是提升奖励，那么奖励越小，反之同理，符合我们调整参数的目标。

### 参数初始化

````python
# 模型相关
s_epoch = 0
total_epoch = 10
s_episode = 0
total_episode = 200
torch.manual_seed(0)
best_score = 0
return_list = []
gamma = 0.98  # 折扣率
learning_rate = 1e-3  # 学习率
env_name = "CartPole-v1"
env = gym.make(env_name)

# 神经网络相关
hidden_dim = 128
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.n
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

agent = REINFORCE(state_dim, hidden_dim, action_dim, learning_rate, gamma, device)
````

### 读取检查点

````python
def read_ckp(CKP_PATH):
    if os.path.exists(CKP_PATH):
        checkpoint = torch.load(CKP_PATH)
        return {
            'epoch': checkpoint.get('epoch', 0), #
            'episode': checkpoint.get('episode', 0),
            #'epsilon': checkpoint.get('epsilon', 1),
            'best_weight': checkpoint.get('best_weight', None),
            'return_list': checkpoint.get('return_list', [])
        }
    else:
        # 创建路径所需的文件夹
        os.makedirs(os.path.dirname(CKP_PATH), exist_ok=True)
        return {
            'epoch': 0,
            'episode': 0,
            'best_weight': None,
            'return_list': []
        }

def save_ckp(data, PATH):
    os.makedirs(os.path.dirname(PATH), exist_ok=True)  # 确保路径存在
    torch.save(data, PATH)
````

````python
CKP_PATH = 'tmp/8_v3_PolicyNet_.pt'
checkpoint_data = read_ckp(CKP_PATH)

s_epoch = checkpoint_data['epoch']
s_episode = checkpoint_data['episode']
best_weight = checkpoint_data['best_weight']
return_list = checkpoint_data['return_list']

if best_weight is not None:
    agent.policy_net.load_state_dict(checkpoint_data['best_weight'])
````

### 训练

````python
for epoch in range(s_epoch, total_epoch):
    with tqdm(total=(total_episode - s_episode), desc='<%d/%d>' % (epoch + 1, total_epoch), leave=False) as pbar:
        for episode in range(s_episode, total_episode):
            episode_reward = 0
            transition_dict = {
                'states': [],
                'actions': [],
                'next_states': [],
                'rewards': [],
                'dones': [],
                'truncated' : [],
            }
            state = env.reset()[0]
            done = truncated = False
            while not (done | truncated):
                action = agent.take_action(state)
                next_state, reward, done, truncated, _ = env.step(action)
                transition_dict['states'].append(state)
                transition_dict['actions'].append(action)
                transition_dict['next_states'].append(next_state)
                transition_dict['rewards'].append(reward)
                transition_dict['dones'].append(done)
                transition_dict['truncated'].append(truncated)
                
                state = next_state
                episode_reward += reward
            return_list.append(episode_reward)
            agent.update(transition_dict)
            
            if episode_reward > best_score:
                best_weight = agent.policy_net.state_dict()
                best_score = episode_reward
            
            if (episode + 1) % 10 == 0:
                pbar.set_postfix({
                    'episode':
                    '%d' % (total_episode * epoch + episode + 1),
                    'recent_return':
                    '%.3f' % np.mean(return_list[-10:])
                })
            
            data_to_save = {
            'epoch': epoch,
            'episode': episode,
            'best_weight': best_weight,
            'return_list': return_list,
            }
            
            save_ckp(data_to_save, CKP_PATH)  
            pbar.update(1)
        
        #rl_utils.picture_return(return_list, 'REINFORCE', env_name, 9)
        s_episode = 0
        
agent.policy_net.load_state_dict(best_weight)  # 应用最佳权重
````

### REINFORCE优缺点：
REINFORCE 算法有效地学习到了最优策略。不过，相比于前面的 DQN 算法，REINFORCE 算法使用了更多的序列，这是因为 REINFORCE 算法是一个在线策略算法，之前收集到的轨迹数据不会被再次利用。此外，REINFORCE 算法的性能也有一定程度的波动，这主要是因为每条采样轨迹的回报值波动比较大，这也是 REINFORCE 算法主要的不足。


### 总结
REINFORCE 算法是策略梯度乃至强化学习的典型代表，智能体根据当前策略直接和环境交互，通过采样得到的轨迹数据直接计算出策略参数的梯度，进而更新当前策略，使其向最大化策略期望回报的目标靠近。

这种学习方式是典型的从交互中学习，并且其优化的目标（即策略期望回报）正是最终所使用策略的性能，**这比基于价值的强化学习算法的优化目标（一般是时序差分误差的最小化）要更加直接**。

REINFORCE 算法理论上是能**保证局部最优的**，它实际上是**借助蒙特卡洛方法**采样轨迹来估计动作价值，这种做法的一大优点是可以得到**无偏的梯度**。

但是，正是因为使用了蒙特卡洛方法，REINFORCE 算法的梯度**估计的方差很大**，可能会造成一定程度上的不稳定，这也是第 10 章将介绍的 Actor-Critic 算法要解决的问题。


## 备注/附录：

#### |什么是梯度？
1. **什么是梯度？**
    - 在多变量函数中，梯度是一个向量，它的方向指向函数增长最快的方向，而它的大小（或模）表示该方向上的函数增长率——**简单地说，梯度告诉我们在哪个方向上函数值上升得最快**
2. **梯度的定义**:
    - 对于一个多变量函数$ f(\mathbf{x})$，其中$\mathbf{x} $ 是一个变量向量$[x_1, x_2, \dots, x_n]$，该函数的梯度定义为所有偏导数的向量：$$ \nabla f(\mathbf{x}) = \left[ \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_n} \right] $$
3. **具体到$\nabla_{\theta} J(\theta)$**:
    - $J(\theta)$表示的是在策略$\pi_{\theta}$下的期望回报
    - $\nabla_{\theta} J(\theta) $是一个向量，其中每个分量是$J$关于$\theta$的某个分量的偏导数
    
---

#### |梯度的含义？
梯度是一个向量，表示多变量函数在给定点的最大增长方向和速率：
1. **方向**：梯度的方向指向函数增长最快的方向，沿着梯度的方向移动，函数的值将最快地增加
2. **大小**：梯度的模表示函数在该方向上的增长率
3. **局部信息**：梯度提供给定点的局部信息，不告诉全局的行为
4. **用于优化**：在梯度下降方法中，我们沿着梯度的负方向更新变量，因为这会减少函数的值

简单来说，梯度为我们提供了一个方向和一个大小，告诉我们在哪个方向上函数增长得最快，以及这个增长是多么迅速。在机器学习和深度学习中，我们使用梯度来更新模型的参数，使损失函数的值最小化或性能度量最大化。

---

#### |为何使用似然比技巧？
在策略梯度中，将$\pi_{\theta}(a \mid s)$转换为含有log的函数，主要原因是数学和计算上的便利性，以及梯度的稳定性：
1. **计算稳定性**：特别是深度神经网络，计算$\nabla_{\theta} \log \pi_{\theta}(a \mid s)$比直接计算$\nabla_{\theta} \pi_{\theta}(a \mid s)$更稳定，这是因为概率$\pi_{\theta}(a \mid s)$可能接近 0 或 1，这在计算梯度时可能会导致数值问题
2. **与其他技术的结合**：似然比形式的策略梯度与某些其他技术，例如**基线**的引入更容易结合，进一步减少方差并提高学习稳定性
3. **似然比技巧**：$$\nabla_{\theta} \log \pi_{\theta}(a \mid s)=\frac{\nabla_{\theta} \pi_{\theta}(a \mid s)}{\pi_{\theta}(a \mid s)}$$

---

#### |$\pi_{\theta}(a \mid s)$是怎么消失的？
从这一步：$\sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} \pi_{\theta}(a \mid s) Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \log \pi_{\theta}(a \mid s)$到这一步：$\mathbb{E}_{\pi_{\theta}}\left[Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \log \pi_{\theta}(a \mid s)\right]$其中$\pi_{\theta}(a \mid s)$是怎么消失的?

这一转换涉及到期望的定义,考虑期望的定义：
$$
\mathbb{E}_{\pi_{\theta}}[f(s, a)] = \sum_{s \in S} \sum_{a \in A} p(s, a) f(s, a)
$$
其中$p(s, a)$是在状态s选择动作a的联合概率，而$f(s, a)$是一个随机变量
现在，回到我们的策略梯度公式。考虑这个部分：
$$
\sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} \pi_{\theta}(a \mid s) Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \log \pi_{\theta}(a \mid s)
$$
当我们乘以$\pi_{\theta}(a \mid s)$和$\nu^{\pi_{\theta}}(s)$时，实际上是在计算状态s和动作a的联合概率。因此，这个公式实际上是一个期望的形式。

---

#### |Categorical分布
Categorical分布是离散概率分布的一种，用于描述具有有限数量可能结果的随机实验的概率分布。每个可能的结果都有一个与之关联的概率：
一个普通的六面骰子有6个可能的结果，每个结果的概率都是1/6
——这可以描述为一个Categorical分布，其中每个结果都有一个与之关联的概率
`torch.distributions.Categorical`类提供了一个Categorical分布的实现
```python
action_dist = torch.distributions.Categorical(probs) 
```

---
