### |DQN-åŸºäºä»·å€¼ & ç­–ç•¥æ¢¯åº¦-åŸºäºç­–ç•¥
Q-learningã€DQN åŠ DQN æ”¹è¿›ç®—æ³•éƒ½æ˜¯**åŸºäºä»·å€¼ï¼ˆvalue-basedï¼‰**çš„æ–¹æ³•ï¼Œè€Œ **åŸºäºç­–ç•¥ï¼ˆpolicy-basedï¼‰** çš„æ–¹æ³•ï¼š
ç­–ç•¥æ¢¯åº¦æ˜¯åŸºäºç­–ç•¥çš„æ–¹æ³•çš„åŸºç¡€ï¼Œæœ¬ç« ä»ç­–ç•¥æ¢¯åº¦ç®—æ³•è¯´èµ·ã€‚

## ç­–ç•¥æ¢¯åº¦ç®—æ³•
ç›®æ ‡ç­–ç•¥$\pi_\theta$æ˜¯ä¸€ä¸ªéšæœºæ€§ç­–ç•¥ï¼Œå¯ä»¥ç”¨ä¸€ä¸ªçº¿æ€§æ¨¡å‹/ç¥ç»ç½‘ç»œæ¥ä¸ºç­–ç•¥å‡½æ•°å»ºæ¨¡ã€‚ç›®æ ‡æ˜¯è¦å¯»æ‰¾ä¸€ä¸ªæœ€ä¼˜ç­–ç•¥å¹¶æœ€å¤§åŒ–è¿™ä¸ªç­–ç•¥çš„æœŸæœ›å›æŠ¥ï¼Œç›®æ ‡å‡½æ•°ä¸ºï¼š
$$
J(\theta)=\mathbb{E}_{s_{0}}\left[V^{\pi_{\theta}}\left(s_{0}\right)\right]
$$
å…¶ä¸­$s_0$ä¸ºåˆå§‹çŠ¶æ€ï¼Œå°†ç›®æ ‡å‡½æ•°Jå¯¹$\theta$æ±‚å¯¼ï¼Œå°±å¯ä»¥ç”¨æ¢¯åº¦ä¸Šå‡æ¥æœ€å¤§åŒ–ç›®æ ‡å‡½æ•°

å¯¹ç›®æ ‡å‡½æ•°æ±‚æ¢¯åº¦ï¼Œå¯ä»¥å¾—åˆ°å¦‚ä¸‹å¼å­:
$$
\begin{aligned} \nabla_{\theta} J(\theta) & \propto \sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \pi_{\theta}(a \mid s) \\ & =\sum_{s \in S} \nu^{\pi_{\theta}}(s) \sum_{a \in A} \pi_{\theta}(a \mid s) Q^{\pi_{\theta}}(s, a) \frac{\nabla_{\theta} \pi_{\theta}(a \mid s)}{\pi_{\theta}(a \mid s)} \\ & =\mathbb{E}_{\pi_{\theta}}\left[Q^{\pi_{\theta}}(s, a) \nabla_{\theta} \log \pi_{\theta}(a \mid s)\right]\end{aligned}
$$
1. **ç›®æ ‡å‡½æ•°**:$J(\theta)$æ˜¯ç­–ç•¥$\pi_{\theta}$çš„æ€§èƒ½åº¦é‡,è¡¨ç¤ºè¯¥ç­–ç•¥çš„æœŸæœ›å›æŠ¥
2. **ç­–ç•¥æ¢¯åº¦**:$\nabla_{\theta} J(\theta)$
3. **çŠ¶æ€è®¿é—®é¢‘ç‡**:$\nu^{\pi_{\theta}}(s)$è¡¨ç¤ºè¯¥ç­–ç•¥ä¸‹è®¿é—®çŠ¶æ€sçš„é¢‘ç‡
4. **åŠ¨ä½œä»·å€¼å‡½æ•°**:$Q^{\pi_{\theta}}(s, a)$è¡¨ç¤ºs,aè·å¾—çš„æœŸæœ›å›æŠ¥
5. **ç­–ç•¥æ¢¯åº¦çš„æ¨å¯¼**:
   - é¦–å…ˆï¼Œæ¢¯åº¦æ˜¯é€šè¿‡æ‰€æœ‰çŠ¶æ€çš„åŠ æƒå’Œæ¥è®¡ç®—çš„ï¼Œæƒé‡æ˜¯æ¯ä¸ªçŠ¶æ€çš„è®¿é—®é¢‘ç‡
   - å¯¹äºæ¯ä¸ªçŠ¶æ€ï¼Œæˆ‘ä»¬è€ƒè™‘æ‰€æœ‰å¯èƒ½çš„åŠ¨ä½œä»·å€¼ï¼Œå¹¶ä¹˜ä»¥ç®—å­
   - æœ€åï¼Œæˆ‘ä»¬å°†è¿™ä¸ªæ±‚å’Œè½¬åŒ–ä¸ºä¸€ä¸ªæœŸæœ›å€¼
   
> æ³¨æ„ï¼šç­–ç•¥æ¢¯åº¦ä¸ºåœ¨çº¿ç­–ç•¥ç®—æ³•!å› ä¸ºä¸Šå¼ä¸­æœŸæœ›$\mathbb{E}_{\pi_{\theta}}$çš„ä¸‹æ ‡æ˜¯$\pi_\theta$ï¼Œå³å¿…é¡»ä½¿ç”¨å½“å‰ç­–ç•¥$\pi_\theta$é‡‡æ ·çš„æ•°æ®æ¥è®¡ç®—æ¢¯åº¦

---

### REINFORCE ç®—æ³•
å¯ä»¥ç”¨å¤šç§æ–¹å¼ä¼°è®¡$Q^{\pi_{\theta}}(s, a)$ï¼Œæ¥ä¸‹æ¥è¦ä»‹ç»çš„ REINFORCE ç®—æ³•ä¾¿æ˜¯é‡‡ç”¨äº†è’™ç‰¹å¡æ´›æ–¹æ³•æ¥ä¼°è®¡ã€‚äºä¸€ä¸ªæœ‰é™æ­¥æ•°çš„ç¯å¢ƒæ¥è¯´ï¼ŒREINFORCE ç®—æ³•ä¸­çš„ç­–ç•¥æ¢¯åº¦ä¸ºï¼š
$$
\nabla_{\theta} J(\theta)=\mathbb{E}_{\pi_{\theta}}\left[\sum_{t=0}^{T}\left(\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}\right) \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)\right]
$$
å…¶ä¸­ï¼ŒTæ˜¯å’Œç¯å¢ƒäº¤äº’çš„æœ€å¤§æ­¥æ•°ã€‚ä¾‹å¦‚ï¼Œåœ¨è½¦æ†ç¯å¢ƒä¸­ï¼ŒT=200

#### |REINFORCEå…·ä½“å®ç°æµç¨‹ï¼š
- åˆå§‹åŒ–ç­–ç•¥å‚æ•°$\theta$
- **for** åºåˆ—$e=1 \rightarrow E$ **do** :
- ----ç”¨å½“å‰ç­–ç•¥$\pi_{\theta}$é‡‡æ ·è½¨è¿¹$\left\{s_{1}, a_{1}, r_{1}, s_{2}, a_{2}, r_{2}, \ldots s_{T}, a_{T}, r_{T}\right\}$
- ----è®¡ç®—å½“å‰è½¨è¿¹æ¯ä¸ªæ—¶åˆ»tå¾€åçš„å›æŠ¥$\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}$ ï¼Œè®°ä¸º $\psi_{t}$
- ----å¯¹$\theta$è¿›è¡Œæ›´æ–°,$\theta=\theta+\alpha \sum_{t}^{T} \psi_{t} \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$
- **end for**

### ç­–ç•¥ç½‘ç»œ
è¿™é‡Œé‡‡ç”¨åœ¨ç¦»æ•£åŠ¨ä½œç©ºé—´ä¸Šçš„softmax()å‡½æ•°æ¥å®ç°ä¸€ä¸ªå¯å­¦ä¹ çš„**å¤šé¡¹åˆ†å¸ƒï¼ˆmultinomial distributionï¼‰**

#### softmax()æ¿€æ´»å‡½æ•°
å°†ä»»æ„å®æ•°å‘é‡è½¬æ¢ä¸ºæ¦‚ç‡åˆ†å¸ƒã€‚ç»™å®šä¸€ä¸ªå®æ•°å‘é‡ï¼ŒSoftmaxå°†ä¸ºæ¯ä¸ªå…ƒç´ åˆ†é…ä¸€ä¸ªä»‹äº0å’Œ1ä¹‹é—´çš„æ¦‚ç‡å€¼ï¼Œä¸”æ‰€æœ‰å…ƒç´ çš„æ¦‚ç‡å€¼ä¹‹å’Œä¸º1ã€‚
$$
\text{Softmax}(z)_i = \frac{e^{z_i}}{\sum_j e^{z_j}}
$$
åœ¨ç­–ç•¥ç½‘ç»œä¸­ï¼Œä½¿ç”¨Softmaxæ˜¯å¾ˆæœ‰æ„ä¹‰çš„ï¼Œå› ä¸ºæˆ‘ä»¬æƒ³è¦è¾“å‡ºçš„æ˜¯ä¸€ä¸ªæ¦‚ç‡åˆ†å¸ƒï¼ˆå³æ¯ä¸ªåŠ¨ä½œçš„æ¦‚ç‡ï¼‰ã€‚Softmaxç¡®ä¿äº†è¾“å‡ºæ˜¯ä¸€ä¸ªæœ‰æ•ˆçš„æ¦‚ç‡åˆ†å¸ƒã€‚

ä½†æ˜¯ï¼ŒReLUä¸é€‚åˆç”¨ä½œç­–ç•¥ç½‘ç»œçš„è¾“å‡ºå±‚çš„æ¿€æ´»å‡½æ•°ï¼Œå› ä¸ºå®ƒä¸ä¼šè¾“å‡ºä¸€ä¸ªæ¦‚ç‡åˆ†å¸ƒã€‚ReLUçš„è¾“å‡ºæ˜¯éè´Ÿçš„ï¼Œä½†å®ƒä»¬çš„å’Œä¸ä¸€å®šæ˜¯1ï¼Œè€Œä¸”å®ƒä»¬ä¹Ÿä¸ä¸€å®šéƒ½åœ¨0å’Œ1ä¹‹é—´

````python
class PolicyNet(torch.nn.Module):
    def __init__(self, state_dim, hidden_dim, action_dim):
        super(PolicyNet, self).__init__()
        self.fc1 = torch.nn.Linear(state_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, action_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        #x = F.relu(self.fc2(x))
        x = self.fc2(x)
        return F.softmax(x, dim=1)  # è¾“å‡ºå„åŠ¨ä½œçš„æ¦‚ç‡
````

### REINFORCEç®—æ³•
æŒ‰ç…§ç®—æ³•å°†æŸå¤±å‡½æ•°å†™ä¸ºç­–ç•¥å›æŠ¥çš„è´Ÿæ•°ï¼Œ$-\sum_{t} \psi_{t} \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$ï¼Œå¯¹$\theta$æ±‚å¯¼å°±å¯ä»¥é€šè¿‡æ¢¯åº¦ä¸‹é™æ¥æ›´æ–°ç­–ç•¥


````python
class REINFORCE:
    def __init__(self, state_dim, hidden_dim, action_dim, learning_rate, gamma,
                 device):
        self.policy_net = PolicyNet(state_dim, hidden_dim,
                                    action_dim).to(device)
        self.optimizer = torch.optim.Adam(self.policy_net.parameters(), lr=learning_rate)  # ä½¿ç”¨Adamä¼˜åŒ–å™¨
        self.gamma = gamma  # æŠ˜æ‰£å› å­
        self.device = device
    def take_action(self, state):  # æ ¹æ®åŠ¨ä½œæ¦‚ç‡åˆ†å¸ƒéšæœºé‡‡æ ·
        state = torch.tensor(state, dtype=torch.float).to(self.device)
        probs = self.policy_net(state.unsqueeze(0))  
        # ---------------ç»™stateåŠ ä¸€ä¸ªç»´åº¦ğŸ‘†
            #ç½‘ç»œé€šå¸¸æœŸæœ›çš„è¾“å…¥æ˜¯ä¸€ä¸ªæ‰¹æ¬¡çš„æ•°æ®ï¼Œå³ä¸º[batch_size, state_dim]
            #å³ä½¿åªæœ‰ä¸€ä¸ªçŠ¶æ€ï¼Œä¹Ÿéœ€è¦è§†ä½œæ‰¹æ¬¡å¤§å°ä¸º1çš„æ‰¹æ¬¡
            #[0.5, 1.2, -0.3]->[[0.5, 1.2, -0.3],] batch_size = 1
            #ä¸ºäº†é«˜æ•ˆåœ°è¿›è¡Œè®¡ç®—ï¼Œé€šå¸¸ä¸€æ¬¡å¤„ç†å¤šä¸ªæ•°æ®æ ·æœ¬
        action_dist = torch.distributions.Categorical(probs)  # æ ¹æ®probsç»™å‡ºçš„æ¦‚ç‡åˆ›å»ºä¸€ä¸ªç±»åˆ«åˆ†å¸ƒ
        #---------------------ä½¿ç”¨Categoricalåˆ†å¸ƒğŸ‘†
            #æ˜¯ç¦»æ•£æ¦‚ç‡åˆ†å¸ƒçš„ä¸€ç§ï¼Œæè¿°å…·æœ‰æœ‰é™æ•°é‡å¯èƒ½ç»“æœçš„éšæœºå®éªŒçš„æ¦‚ç‡åˆ†å¸ƒ
        action = action_dist.sample()  # ä»åˆ›å»ºçš„ç±»åˆ«åˆ†å¸ƒä¸­æŠ½æ ·, å¾—åˆ°æŸä¸ªç±»åˆ«çš„åºå·
        return action.item()

    def update(self, transition_dict):
        return_list = transition_dict['rewards']
        state_list = transition_dict['states']
        action_list = transition_dict['actions']

        G = 0
        self.optimizer.zero_grad() # è®¾å®š0æ¢¯åº¦
        for i in reversed(range(len(return_list))):  # ä»æœ€åä¸€æ­¥ç®—èµ·, å¯¹æ¯ä¸€ä¸ªçŠ¶æ€åŠ¨ä½œè®¡ç®—æŸå¤±
            reward = return_list[i]
            state = torch.tensor(state_list[i], dtype=torch.float).to(self.device)
            action = torch.tensor(action_list[i]).view(-1, 1).to(self.device)  # é€‰æ‹©å½“å‰æ­¥éª¤çš„åŠ¨ä½œ
            
            log_prob = torch.log(
                self.policy_net(state.unsqueeze(0)).gather(1, action))  # ä¾æ®åŠ¨ä½œé€‰æ¦‚ç‡, å¯¹æ•°åŒ–
            
            G = self.gamma * G + reward
            loss = -log_prob * G  # æ¯ä¸€æ­¥çš„æŸå¤±å‡½æ•°
            loss.backward()  # åå‘ä¼ æ’­è®¡ç®—æ¢¯åº¦, åœ¨å¾ªç¯ä¸­ä¸€ç›´ç§¯ç´¯æ¢¯åº¦
        self.optimizer.step()  # æ¢¯åº¦ä¸‹é™, æ¯è½®æ¸¸æˆç»“æŸåè®¡ç®—çš„æ€»æŸå¤±æ¢¯åº¦å…¨éƒ¨æ›´æ–°ä¸€æ¬¡
````

#### updateä¸å…¬å¼çš„å…³ç³»ï¼š
å¦‚ä¸Šupdateä¸å…¬å¼$\nabla_{\theta} J(\theta)=\mathbb{E}_{\pi_{\theta}}\left[\sum_{t=0}^{T}\left(\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}\right) \nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)\right]$çš„å¯¹åº”å…³ç³»ï¼š

> æ³¨æ„ï¼šæ¢¯åº¦æ˜¯é loss.backward()è‡ªåŠ¨è®¡ç®—çš„ï¼Œè¿™æ˜¯é€šè¿‡é“¾å¼æ³•åˆ™å®Œæˆçš„ï¼ï¼ï¼

1. **æœŸæœ›è®¡ç®—**:
   - å…¬å¼ä¸­çš„æœŸæœ›$\mathbb{E}_{\pi_{\theta}}[\cdot]$è¡¨ç¤ºæŒ‰ç­–ç•¥$\pi_{\theta}$ é‡‡æ ·çš„è½¨è¿¹çš„æœŸæœ›ã€‚åœ¨`update`å‡½æ•°ä¸­ï¼Œæ²¡æœ‰æ˜¾å¼åœ°è®¡ç®—æœŸæœ›ï¼Œè€Œæ˜¯é€šè¿‡å®é™…çš„è½¨è¿¹ï¼ˆå³ä»å®é™…çš„ç¯å¢ƒäº¤äº’ä¸­é‡‡æ ·çš„çŠ¶æ€ã€åŠ¨ä½œå’Œå¥–åŠ±åºåˆ—ï¼‰æ¥è¿›è¡Œæ›´æ–°
2. **å¥–åŠ±çš„ç´¯ç§¯**:
   - å…¬å¼ä¸­çš„$\sum_{t^{\prime}=t}^{T} \gamma^{t^{\prime}-t} r_{t^{\prime}}$ è¡¨ç¤ºä»æ—¶é—´æ­¥$t$åˆ°ç»“æŸçš„ç´¯ç§¯æŠ˜æ‰£å¥–åŠ±ã€‚åœ¨`update`å‡½æ•°ä¸­ï¼Œè¿™é€šè¿‡å˜é‡ `G` æ¥è®¡ç®—ï¼Œå®ƒåœ¨æ¯æ¬¡å¾ªç¯è¿­ä»£ä¸­éƒ½ä¼šç´¯ç§¯ï¼ˆæŠ˜æ‰£çš„ï¼‰å¥–åŠ±ã€‚
3. **ç­–ç•¥çš„å¯¹æ•°æ¢¯åº¦**:
   - å…¬å¼ä¸­çš„$\nabla_{\theta} \log \pi_{\theta}\left(a_{t} \mid s_{t}\right)$æ˜¯åœ¨æ—¶é—´æ­¥$t$ï¼Œç»™å®šçŠ¶æ€$s_t$å’Œé‡‡å–çš„åŠ¨ä½œ$a_t$ä¸‹ç­–ç•¥çš„å¯¹æ•°æ¢¯åº¦ã€‚åœ¨`update`å‡½æ•°ä¸­ï¼Œè¿™é€šè¿‡ä»¥ä¸‹ä»£ç æ¥è®¡ç®—ï¼š
     ```python
     log_prob = torch.log(self.policy_net(state.unsqueeze(0)).gather(1, action))
     ```
4. **æŸå¤±å‡½æ•°**:
   - åœ¨ç­–ç•¥æ¢¯åº¦æ–¹æ³•ä¸­ï¼Œæˆ‘ä»¬çš„ç›®æ ‡æ˜¯æœ€å¤§åŒ–æœŸæœ›çš„ç´¯ç§¯å¥–åŠ±ã€‚ä½†æ˜¯ï¼Œç”±äºä¼˜åŒ–å™¨é€šå¸¸æœ€å°åŒ–æŸå¤±ï¼Œæ‰€ä»¥æˆ‘ä»¬ä½¿ç”¨è´Ÿçš„ç­–ç•¥æ¢¯åº¦ä½œä¸ºæŸå¤±ï¼š
     ```python
     loss = -log_prob * G
     ```
5. **æ¢¯åº¦çš„ç´¯ç§¯**:
   - å¯¹äºæ¯ä¸€ä¸ªæ—¶é—´æ­¥ï¼Œéƒ½è®¡ç®—æŸå¤±å¹¶è°ƒç”¨`loss.backward()`ï¼Œè¿™ä¼šç´¯ç§¯æ¢¯åº¦ï¼Œç›´åˆ°æ•´ä¸ªè½¨è¿¹éƒ½å¤„ç†å®Œã€‚
6. **å‚æ•°æ›´æ–°**:
   - ä½¿ç”¨`self.optimizer.step()`ï¼Œæˆ‘ä»¬ä½¿ç”¨ç´¯ç§¯çš„æ¢¯åº¦æ¥æ›´æ–°ç­–ç•¥å‚æ•°ï¼Œå®ç°äº†ç­–ç•¥æ¢¯åº¦çš„æ›´æ–°ã€‚

### âœ…æ³¨æ„ç‚¹
ç­–ç•¥æ¢¯åº¦çš„æŸå¤±ï¼Œä¸æ˜¯ä¸æ­£ç¡®çš„åŠ¨ä½œæ¥å¯¹æ¯”å¾—åˆ°äº¤å‰ç†µï¼Œå› ä¸ºæˆ‘ä»¬äº‹å…ˆä¸çŸ¥é“æ­£ç¡®çš„åŠ¨ä½œï¼Œæ‰€ä»¥æ²¡æœ‰å‚è€ƒï¼Œå› æ­¤å°±åªèƒ½äººä¸ºç¡®å®šä¸€ç§æŸå¤±ï¼Œä¾‹å¦‚ç”¨ç½‘ç»œç»™å‡ºçš„å„ä¸ªåŠ¨ä½œçš„æ¦‚ç‡ï¼Œä¹˜ä»¥å½“å‰å¾—åˆ°çš„å¥–åŠ±ï¼Œå¦‚æœå¥–åŠ±æ¯”è¾ƒä½ï¼Œä¹˜èµ·æ¥ä¹Ÿæ¯”è¾ƒå°ï¼Œæˆ‘ä»¬å¸Œæœ›æå‡å¥–åŠ±ï¼Œæ‰€ä»¥åŠ ä¸ªè´Ÿå·å†æ‰§è¡Œæ¢¯åº¦ä¸‹é™ï¼Œç­‰äºæ˜¯åšäº†æ¢¯åº¦ä¸Šå‡ï¼Œä¹Ÿå°±å¸Œæœ›æå‡å¥–åŠ±ï¼Œå› ä¸ºæ¦‚ç‡è‚¯å®šæ˜¯æ­£çš„ï¼Œæ‰€ä»¥æ˜¯æå‡å¥–åŠ±ï¼Œé‚£ä¹ˆå¥–åŠ±è¶Šå°ï¼Œåä¹‹åŒç†ï¼Œç¬¦åˆæˆ‘ä»¬è°ƒæ•´å‚æ•°çš„ç›®æ ‡ã€‚

### å‚æ•°åˆå§‹åŒ–

````python
# æ¨¡å‹ç›¸å…³
s_epoch = 0
total_epoch = 10
s_episode = 0
total_episode = 200
torch.manual_seed(0)
best_score = 0
return_list = []
gamma = 0.98  # æŠ˜æ‰£ç‡
learning_rate = 1e-3  # å­¦ä¹ ç‡
env_name = "CartPole-v1"
env = gym.make(env_name)

# ç¥ç»ç½‘ç»œç›¸å…³
hidden_dim = 128
state_dim = env.observation_space.shape[0]
action_dim = env.action_space.n
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

agent = REINFORCE(state_dim, hidden_dim, action_dim, learning_rate, gamma, device)
````

### è¯»å–æ£€æŸ¥ç‚¹

````python
def read_ckp(CKP_PATH):
    if os.path.exists(CKP_PATH):
        checkpoint = torch.load(CKP_PATH)
        return {
            'epoch': checkpoint.get('epoch', 0), #
            'episode': checkpoint.get('episode', 0),
            #'epsilon': checkpoint.get('epsilon', 1),
            'best_weight': checkpoint.get('best_weight', None),
            'return_list': checkpoint.get('return_list', [])
        }
    else:
        # åˆ›å»ºè·¯å¾„æ‰€éœ€çš„æ–‡ä»¶å¤¹
        os.makedirs(os.path.dirname(CKP_PATH), exist_ok=True)
        return {
            'epoch': 0,
            'episode': 0,
            'best_weight': None,
            'return_list': []
        }

def save_ckp(data, PATH):
    os.makedirs(os.path.dirname(PATH), exist_ok=True)  # ç¡®ä¿è·¯å¾„å­˜åœ¨
    torch.save(data, PATH)
````

````python
CKP_PATH = 'tmp/8_v3_PolicyNet_.pt'
checkpoint_data = read_ckp(CKP_PATH)

s_epoch = checkpoint_data['epoch']
s_episode = checkpoint_data['episode']
best_weight = checkpoint_data['best_weight']
return_list = checkpoint_data['return_list']

if best_weight is not None:
    agent.policy_net.load_state_dict(checkpoint_data['best_weight'])
````

### è®­ç»ƒ

````python
for epoch in range(s_epoch, total_epoch):
    with tqdm(total=(total_episode - s_episode), desc='<%d/%d>' % (epoch + 1, total_epoch), leave=False) as pbar:
        for episode in range(s_episode, total_episode):
            episode_reward = 0
            transition_dict = {
                'states': [],
                'actions': [],
                'next_states': [],
                'rewards': [],
                'dones': [],
                'truncated' : [],
            }
            state = env.reset()[0]
            done = truncated = False
            while not (done | truncated):
                action = agent.take_action(state)
                next_state, reward, done, truncated, _ = env.step(action)
                transition_dict['states'].append(state)
                transition_dict['actions'].append(action)
                transition_dict['next_states'].append(next_state)
                transition_dict['rewards'].append(reward)
                transition_dict['dones'].append(done)
                transition_dict['truncated'].append(truncated)
                
                state = next_state
                episode_reward += reward
            return_list.append(episode_reward)
            agent.update(transition_dict)
            
            if episode_reward > best_score:
                best_weight = agent.policy_net.state_dict()
                best_score = episode_reward
            
            if (episode + 1) % 10 == 0:
                pbar.set_postfix({
                    'episode':
                    '%d' % (total_episode * epoch + episode + 1),
                    'recent_return':
                    '%.3f' % np.mean(return_list[-10:])
                })
            
            data_to_save = {
            'epoch': epoch,
            'episode': episode,
            'best_weight': best_weight,
            'return_list': return_list,
            }
            
            save_ckp(data_to_save, CKP_PATH)  
            pbar.update(1)
        
        #rl_utils.picture_return(return_list, 'REINFORCE', env_name, 9)
        s_episode = 0
        
agent.policy_net.load_state_dict(best_weight)  # åº”ç”¨æœ€ä½³æƒé‡
````
