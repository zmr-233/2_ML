```python
# 导入需要使用的库,其中numpy是支持数组和矩阵运算的科学计算库,而matplotlib是绘图库
import numpy as np
import matplotlib.pyplot as plt
```

```python

class Bandit:
    def __init__(self,k):
        self.prob = np.random.uniform(size=k)
        self.prob_max_i = np.argmax(self.prob)
        self.prob_max=self.prob[self.prob_max_i]
        self.k=k
        
    def get_k(self,k):
        if np.random.rand() < self.prob[k]:
            return self.prob[k]
        else:
            return 0
    
    def get_max(self):
        return self.prob_max
        
```

```python
np.random.seed(1)
bd = Bandit(10)
print(bd.prob_max)
print(bd.prob_max_i)
```

```python
class Bot:
    def __init__(self,bandit):
        self.bandit = bandit
        self.regret = 0
        self.l_count = np.zeros(self.bandit.k)
        self.l_action = []
        self.l_regret = []
        
    def get_next(self):
        pass
    
    def step(self):
        k = self.get_next()
        self.regret += bandit.get_max() - bandit.get_k(k)
        self.l_regret.append(self.regret)
        self.l_action.append(k)
        self.l_count[k]+=1
        
    def run(self,n):
        for _ in range(n):
            self.step()
            
```

```python
class Next(Bot):
    def __init__(self,bandit,eps,exc):
        super().__init__(bandit)
        self.eps = eps
        self.l_exc = np.array([exc]*self.bandit.k)
        
    def get_next(self):
        if np.random.random() < self.eps:
            k = np.random.randint(0,self.bandit.k)
        else:
            k = np.argmax(self.l_exc)
        r = self.bandit.get_k(k)
        self.l_exc[k] +=1./(self.l_count[k]+1)*(r-self.l_exc[k])
        return k
    
```

```python
def plot_results(solvers, solver_names):
    """生成累积懊悔随时间变化的图像。输入solvers是一个列表,列表中的每个元素是一种特定的策略。
    而solver_names也是一个列表,存储每个策略的名称"""
    for idx, solver in enumerate(solvers):
        time_list = range(len(solver.l_regret))  # Changed solver.regrets to solver.l_regret
        plt.plot(time_list, solver.l_regret, label=solver_names[idx])  # Changed solver.regrets to solver.l_regret
    plt.xlabel('Time steps')
    plt.ylabel('Cumulative regrets')
    plt.title('%d-armed bandit' % solvers[0].bandit.k)  # Changed solvers[0].bandit.K to solvers[0].bandit.k
    plt.legend()
    plt.show()
```

```python
np.random.seed(1)
bandit = Bandit(10)
bots = []
bots_eps = []
for i in range(3):
    tmp1 = (i+1)/10
    tmp2 = Next(bandit, tmp1, 1)
    bots.append(tmp2)
    bots_eps.append('Bot-' + str(i) + ' eps=' + str(tmp1))
```

```python
for i in bots:
    i.run(1000)
```

```python
plot_results(bots,bots_eps)
```
